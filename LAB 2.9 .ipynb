```python
query = "SELECT last_name,COUNT(DISTINCT last_name) AS num_distinct_actors FROM actor GROUP BY last_name"
df = pd.read_sql_query(query, connection)
df
```

```python
##2 Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
```

```python
query = "SELECT *, CASE WHEN DAYOFWEEK(rental_date) IN (1,7) THEN 'weekend' ELSE 'workday' END AS day_type FROM rental"
df = pd.read_sql_query(query, connection)
df
```

```python
## 3 Get all films with ARMAGEDDON in the title.
```

```python
query = "SELECT * FROM film WHERE title LIKE '%%ARMAGEDDON%%'"
df = pd.read_sql_query(query, connection)
df
```

```python
##4 Get 10 the longest films.
```

```python
query = "SELECT title, length FROM film WHERE length IS NOT NULL AND length != 0 ORDER BY length DESC LIMIT 10"
df = pd.read_sql_query(query, connection)
df
```

```python
## 5 How many films include Behind the Scenes content?
```

```python
query = '''SELECT COUNT(film_id) AS total_films_BS 
FROM film WHERE special_features
LIKE "%%Behind the Scenes%%"'''
df=pd.read_sql_query(query,connection)
df
```

```python
##6 Which kind of movies (rating) have a mean duration of more than two hours?
```

```python
query = '''SELECT RATING, (avg(length)>120) as avg_length FROM film group by rating '''
df = pd.read_sql_query(query,connection)
df
```

```python
##7 Rank films by length (filter out the rows that have nulls or 0s in length column). 
##In your output, only select the columns title, length, and the rank.
```

```python
query= '''SELECT title, length, RANK() OVER (ORDER BY length) AS movie_rank
FROM film
WHERE length >0'''
df=pd.read_sql_query(query,connection)
df
```
